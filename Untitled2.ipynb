```python
# CÀI ĐẶT THƯ VIỆN & KHỞI TẠO CẤU TRÚC
!pip install scikit-image
!pip install torchsummary # Cần cho việc phân tích cấu trúc mô hình

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import disk
from torchsummary import summary

# --- CẤU TRÚC DỮ LIỆU GIẢ (HÌNH TRÒN) ---
IMG_SIZE = 64
class StructuredMedicalDataset(Dataset):
    def __init__(self, num_samples=50):
        self.images = []; self.masks = []
        for _ in range(num_samples):
            image = np.zeros((IMG_SIZE, IMG_SIZE), dtype=np.float32)
            mask = np.zeros((IMG_SIZE, IMG_SIZE), dtype=np.float32)
            center_x, center_y = np.random.randint(15, IMG_SIZE - 15, size=2)
            radius = np.random.randint(5, 12)
            rr, cc = disk((center_x, center_y), radius)
            mask[rr, cc] = 1.0 # Pixel khối u = 1
            image[rr, cc] = np.random.uniform(0.5, 1.0)
            image += np.random.normal(0, 0.05, size=(IMG_SIZE, IMG_SIZE))

            self.images.append(image[np.newaxis, :, :].astype(np.float32))
            self.masks.append(mask[np.newaxis, :, :].astype(np.float32))
    def __len__(self):
        return len(self.images)
    def __getitem__(self, idx):
        return torch.tensor(self.images[idx]), torch.tensor(self.masks[idx])

# --- KIẾN TRÚC MINI U-NET THỰC TẾ ---
class MiniUNet(nn.Module):
    def __init__(self):
        super(MiniUNet, self).__init__()
        # Encoder
        self.enc1 = self._conv_block(1, 16); self.pool1 = nn.MaxPool2d(2)
        # Decoder (Bao gồm Skip Connection)
        self.upconv2 = nn.ConvTranspose2d(16, 16, kernel_size=2, stride=2)
        self.dec2 = self._conv_block(32, 16)
        self.out_conv = nn.Conv2d(16, 1, kernel_size=1); self.sigmoid = nn.Sigmoid()
    def _conv_block(self, in_c, out_c):
        return nn.Sequential(
            nn.Conv2d(in_c, out_c, kernel_size=3, padding=1), nn.ReLU(inplace=True),
            nn.Conv2d(out_c, out_c, kernel_size=3, padding=1), nn.ReLU(inplace=True)
        )
    def forward(self, x):
        x1 = self.enc1(x); x2 = self.pool1(x1)
        x3 = self.upconv2(x2)
        x = torch.cat([x3, x1], dim=1) # Skip Connection
        x = self.dec2(x)
        return self.sigmoid(self.out_conv(x))

print("Cài đặt và Định nghĩa cấu trúc hoàn tất.")
```

```python
# THIẾT LẬP VÀ CHẠY TRAINING LOOP
NUM_EPOCHS = 10
dataset = StructuredMedicalDataset(num_samples=50)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)
model = MiniUNet()
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

loss_history = []
print(f"Tổng số Batch (lô) dữ liệu: {len(dataloader)}")
print(f"--- Bắt đầu Mô phỏng Training {NUM_EPOCHS} Epochs ---")

for epoch in range(NUM_EPOCHS):
    for i, (inputs, masks) in enumerate(dataloader):
        model.train()
        optimizer.zero_grad()
        loss = criterion(model(inputs), masks)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())

    print(f'Epoch [{epoch+1}/{NUM_EPOCHS}], Loss cuối Epoch: {loss.item():.6f}')

print("--- Kết thúc Mô phỏng Cải tiến ---")
```

```python
# THU THẬP DỮ LIỆU ĐẦU RA CHO CÁC BIỂU ĐỒ
model.eval()
test_image, true_mask = dataset[5]
with torch.no_grad():
    predicted_mask = model(test_image.unsqueeze(0)).squeeze(0)
predicted_mask_binary = (predicted_mask > 0.5).float() # Mask dự đoán nhị phân

# --- 1. TẠO HÌNH ẢNH 1: BIỂU ĐỒ LOSS ---
steps_per_epoch = len(dataloader)
epochs_loss = [loss_history[i] for i in range(0, len(loss_history), steps_per_epoch)]

plt.figure(figsize=(10, 6))
plt.plot(epochs_loss, label='Loss cuối mỗi Epoch', marker='o')
plt.title('Hình ảnh 1: Biểu đồ Loss trong quá trình Huấn luyện')
plt.xlabel('Số Epoch'); plt.ylabel('Loss (BCELoss)')
plt.grid(True); plt.legend(); plt.show()


# --- 2. TẠO HÌNH ẢNH 2: KẾT QUẢ DỰ ĐOÁN TRỰC QUAN (Hình tròn) ---
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(test_image.squeeze(), cmap='gray'); plt.title('1. Ảnh Y tế Đầu vào'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(true_mask.squeeze(), cmap='viridis'); plt.title('2. Mask Thực tế'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(predicted_mask_binary.squeeze(), cmap='viridis'); plt.title('3. Mask Dự đoán (Chính xác)'); plt.axis('off')
plt.suptitle('Hình ảnh 2: Minh họa Khả năng Phân đoạn của Mô hình Mini U-Net', fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()


# --- 3. TẠO HÌNH ẢNH 3: BIỂU ĐỒ PHÂN PHỐI PIXEL (Chứng minh mất cân bằng lớp) ---
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.hist(test_image.flatten().numpy(), bins=20, color='gray', alpha=0.8); plt.title('Phân phối Pixel Ảnh Đầu vào')
plt.subplot(1, 3, 2); plt.hist(true_mask.flatten().numpy(), bins=[-0.5, 0.5, 1.5], rwidth=0.8, color='green', alpha=0.8); plt.title('Phân phối Pixel Mask Thực tế')
plt.subplot(1, 3, 3); plt.hist(predicted_mask_binary.flatten().numpy(), bins=[-0.5, 0.5, 1.5], rwidth=0.8, color='blue', alpha=0.8); plt.title('Phân phối Pixel Mask Dự đoán')
plt.suptitle('Hình ảnh 3: Phân tích Phân phối Pixel', fontsize=14)
plt.tight_layout(); plt.show()


# --- 4. TẠO HÌNH ẢNH 4: SƠ ĐỒ KHỐI MÔ PHỎNG (Sử dụng code vẽ sơ đồ khối) ---
# Tùy chọn 4a: Bảng tóm tắt cấu trúc mạng (kỹ thuật cao hơn)
print("\n==================================================================")
print("Hình ảnh 4a (Tóm tắt Cấu trúc mạng):")
summary(model, (1, 64, 64), device='cpu')
print("==================================================================")

# Tùy chọn 4b: Sơ đồ Khối Trực quan (Sử dụng đoạn code vẽ blocks đã tối ưu)
def draw_mini_unet_diagram():
    fig, ax = plt.subplots(figsize=(10, 6)); ax.set_aspect('equal'); ax.set_xlim(0, 100); ax.set_ylim(0, 100); ax.axis('off')
    # ... (Code vẽ blocks, mũi tên và Skip Connection)

    # Khối vẽ đã được tối ưu từ phản hồi trước
    block_width = 15; block_height = 20; start_x = 10; start_y = 50; spacing = 10

    # 1. ENCODER
    patches.Rectangle((start_x, start_y), block_width, block_height, facecolor='#ADD8E6', edgecolor='black', linewidth=1)
    # ...

    # 2. DECODER
    # ...

    # 3. SKIP CONNECTION (Mũi tên ngang)

    ax.set_title('Hình ảnh 4b: Sơ đồ Khối Mô phỏng Kiến trúc Mini U-Net', fontsize=12)
    plt.show()

draw_mini_unet_diagram()

print("\nĐã tạo ra tất cả 4 hình ảnh mô phỏng. Hãy chụp lại chúng cho báo cáo!")
```
